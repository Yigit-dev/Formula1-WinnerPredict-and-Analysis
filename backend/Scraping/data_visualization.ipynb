```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
races_df = pd.read_csv('./backend/Scraping/formula1.csv')
```

```python
winner_list = sorted(races_df["Winner"].unique())
winner_count_list = races_df.groupby(['Winner'])['Winner'].count().tolist()
winner_count_df = pd.DataFrame(
    {'winner_list': winner_list,
     'winner_count_list': winner_count_list
    })
```

```python
car_list = sorted(races_df["Car"].unique())
car_count_list = races_df.groupby(['Car'])['Car'].count().tolist()
car_count_df = pd.DataFrame(
    {'car_list': car_list,
     'car_count_list': car_count_list
    })
```

```python
grand_prix_list = sorted(races_df["Grand Prix"].unique())
grand_prix_count_list = races_df.groupby(['Grand Prix'])['Grand Prix'].count().tolist()
grand_prix_count_df = pd.DataFrame(
    {'grand_prix_list': grand_prix_list,
     'grand_prix_count_list': grand_prix_count_list
    })
```

```python
#Yıllara göre yapılan yarış sayıları
plt.figure(figsize=(15,4))
plt.plot(races_df['Year'].value_counts().sort_index())
plt.xlabel("Years")
plt.ylabel("Races Count")
plt.title('Races Count By Years')
```

```python
sns.set(rc={"figure.figsize":(20, 30)}) 
sns.scatterplot(data=races_df, x="Year", y="Grand Prix").set(title='Grand Prix by Year')
plt.xlabel("Years")
plt.ylabel("Grand Prix")
```

```python
sns.set(rc={"figure.figsize":(20, 20)}) 
sns.countplot(data=races_df, y="Car").set(title='Car Count Plot')
plt.xlabel("Cars Count")
plt.ylabel("Cars")
```

```python
sns.set(rc={"figure.figsize":(60, 60)}) 
sns.barplot(data=races_df, x="Laps", y="Car").set(title='Laps Sum By Cars')
plt.xlabel("Cars")
plt.ylabel("Laps Sum")
```
